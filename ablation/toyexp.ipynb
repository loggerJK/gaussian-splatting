```python
# Read Image
import cv2  # OpenCV
img_path = '/mnt/sda/home/cvlab03/project/dataset/colmap/bicycle/images/_DSC8681.JPG'
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
img.shape
```

```python
# Downscale Image
downscale = 8 # 1/8 downscale
img = cv2.resize(img, (img.shape[1]//downscale, img.shape[0]//downscale))
```

```python
# Blur Image
import torch
import numpy as np
size = 31
img = cv2.GaussianBlur(img, (size,size), 0)
# img = cv2.GaussianBlur(img, (size,size), 0)

# img = cv2.blur(img, (size, size))
```
